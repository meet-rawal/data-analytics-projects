```python
#1. Program to print maximum among two numbers.
num1=int(input("Enter a no:"))
num2=int(input("Enter a no:"))
if num1>num2:
  print("Maximum number is:",num1)
else:
  print("Maximum number is:",num2)
```

```python
#2. Program to print maximum among three numbers.
a=int(input("Enter a no:"))
b=int(input("Enter a no:"))
c=int(input("Enter a no:"))
if a>b and a>c:
  print(f"{a} is maximum")
elif b>a and b>c:
  print(f"{b} is maximum")
else:
  print(f"{c} is maximum")
```

```python
# 7. Program to check whether a year is a leap year or not.
year = int(input("Enter a year: "))

if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print(year, "is a leap year")
else:
    print(year, "is not a leap year")
```

```python
# 9. Program to check whether a given input is the alphabet or not.
a=input("Enter a input:")
if a.isalpha():
  print("input",a,"is a alphabet...")
else:
  print("Input",a,"is not a alphabet..")
```
