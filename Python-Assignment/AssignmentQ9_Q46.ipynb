```python
#24. Accept the kilometer covered and calculate the bill according to the following criteria.
km = float(input("Enter kilometers covered: "))
if km <= 10:
    bill = km * 11
elif km <= 100:
    bill = 10 * 11 + (km - 10) * 10
else:
    bill = 10 * 11 + 90 * 10 + (km - 100) * 9
print("Total Bill: Rs", bill)

```

```python
#25. Write a program to input electricity unit charges and calculate total electricity bill according to the given condition:
# For first 50 units Rs. 0.50/unit
# For next 100 units Rs. 0.75/unit
# For next 100 units Rs. 1.20/unit
# For unit above 250 Rs. 1.50/unit
# An additional surcharge of 20% is added to the bill.

units=int(input("Enter Electricity units: "))
if units<=50:
  total=units*0.50
elif units<=150:
  total=50*0.50+(units-50)*0.75
elif units<=250:
  total=50*0.50+100*0.75+(units-150)*1.20
else:
  total = 50 * 0.50 + 100 * 0.75 + 100 * 1.20 + (units - 250) * 1.50
total+=0.50
print("Total Electricity Bill : ",total)
```

```python
#36.  Python program to check a given number is Armstrong or not using a for loop.
num = int(input("Enter a number: "))
temp = num
sum = 0
for digit in str(temp):
    sum += int(digit) ** 3

if num == sum:
    print("It is an Armstrong Number")
else:
    print("It is not an Armstrong Number..")

```

```python
# 37.  Python program to print all Armstrong numbers between 1 to n using a for loop.
n = int(input("Enter range: "))
for num in range(1, n + 1):
    power = len(str(num))
    s = 0
    for d in str(num):
        s += int(d) ** power
    if s == num:
        print(num)
```
